```python
import requests
import os
from bs4 import BeautifulSoup as bs 
import pandas as pd
import urllib.parse
import time


geo_ids={
    "Saudi Arabia":"100459316",
    "Egypt":"106155005",
    "United Arab Emirates":"104305776"
}

job_name=input("Enter Job Name:")
parse_job_name=urllib.parse.quote(job_name)

print(parse_job_name)

job_country=input("Enter Job Country:")
geo_id=geo_ids.get(job_country)

print(geo_id)

url=f'https://www.linkedin.com/jobs/search/?geoId={geo_id}&keywords={parse_job_name}&origin=JOB_SEARCH_PAGE_SEARCH_BUTTON&refresh=true'

response=requests.get(url)
soup=bs(response.content,'html.parser')
soup
```

```python
containers=soup.find_all("div",class_="base-card relative w-full hover:no-underline focus:no-underline base-card--link base-search-card base-search-card--link job-search-card")

job_titles,companies,companies_links,locations,publish_time=[],[],[],[],[]
for container in containers:
    job_title=container.find("h3",class_="base-search-card__title")
    job_titles.append(job_title.text.strip())
    
    company=container.find("h4",class_="base-search-card__subtitle")
    companies.append(company.text.strip())

    company_links=container.find("a",class_="hidden-nested-link")
    companies_links.append(company_links.get('href'))

    location=container.find("span",class_="job-search-card__location")
    locations.append(location.text.strip())

    time=container.find("time")
    publish_time.append(time.text.strip())

print(job_titles)
```

```python

df=pd.DataFrame({'job_titles':job_titles,'companies':companies,'companies_links':companies_links,'location':location,'publish_time':publish_time})
```

```python





file_path=f"{job_name}_jobs.csv"

if not os.path.exists(file_path):
    df.to_csv(file_path,index=False,encoding='utf-8-sig')

else:
    df_old=pd.read_csv(file_path,encoding='utf-8-sig')
    df_new=df[~df['companies_links'].isin(df_old['companies_links'])]

    if not df_new.empty:
        df_new.to_csv(file_path,index=False,encoding='utf-8-sig',mode='a',header=False)


#new_data = pd.DataFrame({
#    'job_titles': job_titles,
#    'companies': companies,
#    'companies_links': companies_links,
#    'location': locations,
#    'publish_time': publish_time
#})

# If file exists → read old data
#if os.path.exists(file_name):
#    old_data = pd.read_csv(file_name)
#    
#    # Concatenate old + new
#    df = pd.concat([old_data, new_data], ignore_index=True)
#    
#    # Remove duplicates if needed (based on title + company link for example)
#    df.drop_duplicates(subset=['job_titles', 'companies_links'], inplace=True)
#else:
#    df = new_data

# Save updated file
#df.to_csv(file_name, index=False, encoding='utf-8-sig')
```
